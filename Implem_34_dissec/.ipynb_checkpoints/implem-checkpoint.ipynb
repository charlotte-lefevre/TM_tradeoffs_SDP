## Construire les bonnes listes

Il nous faut diviser la matrice H en bons morceaux de manière à 
* Partitionner celle-ci en 64
* Avoir le bon nombre de solutions

Le problème etant que (k+l)/64 vaut 6.45 ..
* Si on a  une sous matrice de taille 6,alors 2\*\*6 = 64 < 82
* Si on a une sous matrice de taille 7 alors 2\*\*7 > 64

on voit que $k+l = 35*6+29*7$ 

on peut imaginer les listes suivantes :

$6 6 6 7 ll 6 6 6 7 ll 6 6 6 7 ll ( 6 6 7 7 ll 6 6 7 7 ll .. ll 6 6 7 7 ) $

ce qu'il faut comprendre : pour les trois premiers on a 3 6 et 1 7 et pour le reste c'est moit moit

quant au nombre d'élements, On va regarder combien en faut-il, sahant qu'on aimerait quelque chose le plus équilibré possible. Code qui permet de voir ca : on a 64 elements dans chaque 6 sousmatrice, et on choisit dans nb_submatrix le nombre qu'on va prendre dans chaque 7 sous matrice. On le fait augmenter jusqu'à ce qu'on obtienne 800 solutions


```python
nb_submatrix = 108

sisisise = (64.*64*64*nb_submatrix)/3**12
sisisese = (64.*64*nb_submatrix*nb_submatrix)/3**12
trois_sisisise = (sisisise*sisisise*sisisise*sisisese)/3**12
moit_moit = sisisese**4/3**12
final = (trois_sisisise*moit_moit**3)/3**10

print(final)
```

94 parait donc le meilleur compromis

### Rentrer dans une dissection

Ici, tout est multiple de 3, on a pas de problèmes. Par contre, si on devait fuseionner sur, disons 11 symboles, voici une idée de marche à suivre

Pour la premiere et la deuxieme dissection, il nous faut fusionner sur 11 symboles. Celui-ci etant censé etre un mutliple de 3, il va falloir raisonner autrement : 
11 = 4 + 4 + 3. On peut donc explicitement fusionner sur 4 symboles au premier etage, puis ensuite d'occuper des 4+3 autres dans la boucle.
On s'attend à des listes intermédiaires (apres premeire fusion donc) de : 
* Dissection 1 56\*56/3\*\*4 = 38.7
* Dissection 2 55.5\*55.5/3\*\*4 = 38.0

Ainsi, l'idée si on a une contrained e f bits, et que 3 ne divise pas ces f bits, on va prendre le plus petit entier multiple de 3 superier à f, disons 3pf et si 3pf = 3 j alors on fusionne au premiere etage sur j bits. Cela nous permet de ne pas avoir des listes plus grosses que necessaire.

Ainsi, on voit qu'on a besoin d'une nouvelle dissection, où on customise la target sur laquelle on fusionne a chaque etage

# Une nouvelle version de dissection
Suite au constat que la concatenation de listes coute super cher, on a créée unenouvelle fonction "four_dissection_targets_matrix", qui ne sert plus des listes, mais où les vecteurs sont remplacés par des amtrices à une seule colone.
Testons avce n = 560

```python
n = 560
l = 34
k = 379
cardinal_list_6 = 2**6
cardinal_list_7 = 108

target_11 = 4
target_12 = 8

len_number_solutions = []


for _ in range (0,200):
    Lists_level_2 = []
    Lists_level_3 = []
    index_H = 0

    H = random_matrix(F3, l, k+l)
    s = random_matrix(ring = F3, nrows = l, ncols = 1)

    ####################################
    ############# LEVEL 1 ##############
    ####################################
    print("\n\n LEVELLL 1\n\n")
    print("Expected length 53.3 for three first ones otherwise / 92 ")
    for i in range(0,4**3,4):
    # Build 4 lists and do a 4-dissection
    # The H[ugly indexes] represent the columns of H upon which we should do linear combinations for each list
        index_H_tmp = index_H
        if (i < 4*3): # Alors on est dans les 6667:
            L1 = build_list_matrix(H[:,index_H : index_H+6], i, s,cardinal_list_6)
            L2 = build_list_matrix(H[:,index_H+6 : index_H+12], i+1, s,cardinal_list_6)
            L3 = build_list_matrix(H[:,index_H+12 : index_H+18], i+3, s,cardinal_list_6)
            L4 = build_list_matrix(H[:,index_H+18 : index_H+18+7], i+4, s,cardinal_list_7)
            index_H = index_H + 18+7
        else:
            L1 = build_list_matrix(H[:,index_H : index_H+6], i, s,cardinal_list_6)
            L2 = build_list_matrix(H[:,index_H+6 : index_H+12], i+1, s,cardinal_list_6)
            L3 = build_list_matrix(H[:,index_H+12 : index_H+12+7], i+3, s,cardinal_list_7)
            L4 = build_list_matrix(H[:,index_H+12+7 : index_H+12+14], i+4, s,cardinal_list_7)
            index_H = index_H + 12+14


        L1234 = four_dissection_targets_matrix(L1,L2,L3,L4,l, 0,target_11,target_12)

        check_syndrome = (i==0)
        #oracle_test_four_list_tree(H[:,index_H_tmp:index_H],L1234,s, 0,target_11+target_12, check_syndrome)

        Lists_level_2.append(L1234)
        print("len list " + str(len(L1234)))

    # Just for debug, to remove
    assert(len(Lists_level_2) == 4**2)

    ####################################
    ############# LEVEL 2 ##############
    ####################################
    print("\n\n LEVELLL 2\n\n")
    print("Expected length 25.6 for first one or 122.9")

    # Building the lists and doing 4-dissection
    for i in range(0,4**2,4):
        # Build 4 lists and do a 4-dissection
        L1 = Lists_level_2[i]
        L2 = Lists_level_2[i+1]
        L3 = Lists_level_2[i+2]
        L4 = Lists_level_2[i+3]

        L1234 = four_dissection_targets_matrix(L1,L2,L3,L4,l,target_12+target_11, target_11,target_12)

        #check_syndrome = (i==0)
        #oracle_test_four_list_tree(Htmp,L1234,s, 0,(target_11+target_12)*2, check_syndrome)

        print("len list " + str(len(L1234)))

        Lists_level_3.append(L1234)

    # Just for debug, to remove  
    assert(len(Lists_level_3) == 4)

    ####################################
    ############# LEVEL 2 ##############
    ####################################
    print("\n\n LEVELLL 3\n\n")
    print("Expected length 804 ")

    L1 = Lists_level_3[0]
    L2 = Lists_level_3[1]
    L3 = Lists_level_3[2]
    L4 = Lists_level_3[3]

    L = four_dissection_targets_matrix(L1,L2,L3,L4,l, (target_12+target_11)*2, target_11, l, True)
    List_L = L[matrix(F3, nrows=l,ncols=1,immutable = True)]
    print("Number of solutions " + str(len(List_L)))

    print("end")
    print("Theoretical expected number of solutions " + str(RR(3**(2*l/11))))
    
    len_number_solutions.append(len(List_L))
    

print("Checking correctness of results for last iteration .. ")
for l in List_L:
    if (H*l != s):
        print("Ne marche pas ..")
print("If no nasty message was printed, then everything works well")


print("END \n\n Average numberof solutions : " + str(RR(mean(len_number_solutions))))
```
