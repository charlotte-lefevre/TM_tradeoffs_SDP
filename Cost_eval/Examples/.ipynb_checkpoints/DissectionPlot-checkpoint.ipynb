Now, we can display the different regimes of dissections

```python
plt.figure(figsize = (10,7))

r = 300
Re = 0.060835

M,T, colors = dissection(r, R, Re,W, nbMem=100)
plt.scatter(M,T, s=30, marker = 'o', label = str(r)+"-dissection with l = " + str(round(Re,6)) + " x n",c=colors)

plt.legend()

```
