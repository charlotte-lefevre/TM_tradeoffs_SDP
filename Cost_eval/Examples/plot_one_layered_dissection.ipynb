```python
import matplotlib.pyplot as plt
import numpy as np
import math
import sys

sys.path.insert(1, '../')
load("../Wave_param.py")
from kTree import *
import layered_dissections
```

```python
coeff_slope = 2
h = 3 ; r = 4
npoints = 400

Remin = 0.001 ;Remax = 0.2

plt.figure(figsize=(10,7))
StateofTheArtRef = "[4]"
plt.xlabel("log2(M)/n"); plt.ylabel("log2(T)/n")


# Set x,y limits
axes = plt.gca()
axes.set_ylim([0.01*math.log(3,2),0.04])
axes.set_xlim([0.0*math.log(3,2),0.0120*math.log(3,2)])

plt.scatter(0.0176, 0.0176, label=StateofTheArtRef+"'s algorithm", s=40, marker="d",c ="silver")


(memories, times,cols, Res) = layered_dissections.Apply(r, h, R,W, Remin=Remin, Remax=Remax, nbRes=npoints)
plt.scatter(memories, times, s = 3, c=cols, label=str(h)+"-"+str(r) +" dissection")

Ms = np.linspace(0,0.0175,200)
TMM = [coeff_slope*m for m in Ms]
plt.plot(Ms, TMM, "slateblue", label="T=M**" +str(coeff_slope), linewidth=0.4)


plt.legend()
plt.title("Different complexitites - log2")
```
