# Aim : determine which layered dissections are interesting

We want the best possible trade-offs for the layered dissection. Here, we graphically show the layered dissections which are not interesting for us.

Denote by $h$ the number of levels of the considered layered dissection.

Color code:
* Red : we want to show that the considered algorithm is not interesting
* Black : the interesting algortihms with this particular $h$
* Gray: interesting layred dissections with $h' < h$

```python
import sys
import matplotlib.pyplot as plt
sys.path.insert(1, '../')

import numpy as np
import math
import matplotlib.pyplot as plt


import dissection_optimize
import layered_dissections

load("../Wave_param.py")
```

# Two levels (h=2)
## Show that this is not interesting to go beyond a 30-dissection and below a 7-dissection

```python
plt.figure(figsize = (15,10))
axes = plt.gca()
axes.set_ylim([0.01*math.log(3,2),0.04])
axes.set_xlim([0.0*math.log(3,2),0.0120*math.log(3,2)])

Dissections_two  = range(7,30)
h = 2

Remin = 0.002
Remax = 0.07

for X in Dissections_two:
        (memories, times,_, _) = layered_dissections.Apply(X, h, R,W, Remin=Remin, Remax=Remax)
        plt.scatter(memories, times, c = "black")
        


Dissections_two_not_interesting  = np.append(range(2,7), range(30,50,2))
for X in Dissections_two_not_interesting:
        (memories, times,_, _) = layered_dissections.Apply(X, h, R,W, Remin=Remin, Remax=Remax)
        plt.scatter(memories, times, c = "red", s=2)
        
       
plt.vlines(0.0006, 0,0.05)
plt.hlines(0.0006*50, 0, 0.0006, "gray")
```

The blue line represents a memory of 0.001. On the left of this line, the big dissections are slightly better than dissections in the range(7,30), but the points correspond to $T \geq M^{50}$, with a neglectable advantage

From now, all dissections with $h=2$ are displayed with gray color
